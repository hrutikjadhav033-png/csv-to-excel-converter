```python
# CSV → Excel Converter
# Author: Hrutik Jadhav

import argparse
import pandas as pd
import logging
import os

# Logging setup
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def clean_data(df):
    logging.info("Cleaning data...")

    df = df.fillna("")
    df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]

    for col in df.columns:
        try:
            df[col] = pd.to_datetime(df[col])
        except:
            pass

    return df


def convert_csv_to_excel(input_file, output_file):
    if not os.path.exists(input_file):
        logging.error("❌ Input file does not exist!")
        return

    try:
        df = pd.read_csv(input_file)
        logging.info("✅ CSV file loaded successfully")

        df = clean_data(df)

        df.to_excel(output_file, index=False, engine='openpyxl')
        logging.info(f"✅ Excel file saved at: {output_file}")

    except Exception as e:
        logging.error(f"❌ Error processing file: {e}")


if __name__ == "__main__":
    try:
        parser = argparse.ArgumentParser(description="CSV to Excel Converter")

        parser.add_argument("-i", "--input", help="Input CSV file path")
        parser.add_argument("-o", "--output", help="Output Excel file path")

        args, unknown = parser.parse_known_args()

        if args.input and args.output:
            convert_csv_to_excel(args.input, args.output)
        else:
            print("📌 Running with your file...")
            convert_csv_to_excel(
                "C:/Users/HP/Downloads/dataset_2191_sleep.csv",
                "sleep_output.xlsx"
            )

    except Exception as e:
        print(f"Error: {e}")
```
